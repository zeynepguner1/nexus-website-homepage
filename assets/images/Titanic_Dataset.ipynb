**Normal Dağılıma Yakın mı?**

Bu verilere bakıldığında, ortalama (29.7) ve medyan (28) birbirine yakın, ancak hala bazı farklılıklar var.

Standart sapma (14.5), yaşların geniş bir dağılıma sahip olduğunu ve çok genç ve çok yaşlı yolcuların da bulunduğunu gösteriyor.

Min ve Max değerleri, uç yaşları (bebekler ve yaşlılar) içerdiğinden, verinin çarpık bir dağılım gösterdiğini düşündürebilir.

Verilerin çarpık olduğunu gözlemliyoruz, medyan daha iyi bir dolgu yöntemi olacaktır. Çarpık dağılımlarda medyan, ortalama yerine daha sağlam bir değer sağlar.

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
#Embarked sütununda boş değerler var.
#Kategorik veri sütunlarında mod genellikle daha anlamlı bir sonuç elde etmemizi sağlar.
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
# Yaş değişkeninin dağılımı
plt.figure(figsize=(8,4))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Yaş Dağılımı')
plt.show()
```

```python
# Ücret değişkeninin dağılımı
plt.figure(figsize=(8,4))
sns.histplot(df['Fare'], bins=30, kde=True)
plt.title('Ücret Dağılımı')
plt.show()
```

```python
#Log dönüşümü yapabiliriz. Böylece büyük ücretler sıkıştırılır ve daha normal bir dağılım elde edilir.
df['Fare'] = np.log1p(df['Fare'])  # log(1+x) dönüşümü
```

```python
sns.histplot(df['Fare'], bins=30, kde=True)
plt.title('Log Dönüşümü Sonrası Ücret Dağılımı')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x=df['Fare'])
plt.title('Log Dönüşümü Sonrası Fare Boxplot')
plt.show()
```

```python
# IQR hesapla
Q1 = df['Fare'].quantile(0.25)  # 1. çeyrek
Q3 = df['Fare'].quantile(0.75)  # 3. çeyrek
IQR = Q3 - Q1  # IQR hesapla

# Alt ve üst sınırları belirle
alt_sinir = Q1 - 1.5 * IQR
ust_sinir = Q3 + 1.5 * IQR

# Uç değerleri kontrol et
outliers = df[(df['Fare'] < alt_sinir) | (df['Fare'] > ust_sinir)]
print(f'Uç Değer Sayısı: {len(outliers)}')
```

```python
# Uç değerleri filtreleyerek veri setinden çıkar
df = df[(df['Fare'] >= alt_sinir) & (df['Fare'] <= ust_sinir)]

# Temizleme sonrası tekrar uç değer sayısını kontrol et
outliers_after = df[(df['Fare'] < alt_sinir) | (df['Fare'] > ust_sinir)]
print(f'Temizleme Sonrası Uç Değer Sayısı: {len(outliers_after)}')
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x=df['Fare'])
plt.title('Uç Değer Temizliği Sonrası Fare Boxplot')
plt.show()
```

```python
sns.histplot(df['Fare'], bins=30, kde=True)
plt.title('Uç Değer Temizliği Sonrası Ücret Dağılımı')
plt.show()
```
